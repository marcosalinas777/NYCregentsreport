```python
#add one more column to know borough of schools
df['Borough']=df['School DBN'].str[2]
borough_mapping={
'M':'Manhattan',
'X':'Bronx',
'K':'Broolklyn',
'Q':'Queens',
'R':'Staten Island'
}
df['Borough']=df['Borough'].map(borough_mapping)
df
```

```python
#how many English and geometry regents exam where taken in different boroughs
english_geometry_data = df[df['Regents Exam'].isin(['English', 'Geometry'])]
```

```python
exams_taken_by_borough = english_geometry_data.groupby(['Borough', 'Regents Exam']).size().reset_index(name='Number of Exams')
exams_taken_by_borough
```

```python
#visualization of how many English and geometry regents exam where taken in different boroughs
pivot_table = exams_taken_by_borough.pivot(index='Borough', columns='Regents Exam', values='Number of Exams')
```

```python
pivot_table.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Borough')
plt.ylabel('Number of Exams Taken')
plt.title('English and Geometry Regents Exams Taken in Different Boroughs')
plt.legend(title='Regents Exam')
plt.xticks(rotation=0)
plt.show()
```

```python
#how many 'Integrated Algebra' regents exam where taken in different boroughs
integratedalgebra_data = df[df['Regents Exam'].isin(['Integrated Algebra'])]
```

```python
integratedalgebra_data
```

```python
exams_taken_by_borough_integratedalgebra = integratedalgebra_data.groupby(['Borough', 'Regents Exam']).size().reset_index(name='Number of Exams')
exams_taken_by_borough_integratedalgebra
```

```python
#visualization of how many integrated algebra regents exam where taken in different boroughs
pivot_table = exams_taken_by_borough_integratedalgebra.pivot(index='Borough', columns='Regents Exam', values='Number of Exams')
pivot_table.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Borough')
plt.ylabel('Number of Exams Taken')
plt.title('Integrated Algebra Regents Exams Taken in Different Boroughs')
plt.legend(title='Regents Exam')
plt.xticks(rotation=0)
plt.show()
```

```python
#visualization distribution of mean scores
plt.figure(figsize=(10,6))
sns.histplot(df['Mean Score'], bins=0, kde=True)
plt.title('Distribution of Mean Scores')
plt.xlabel('Mean Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example 7: Heatmap - Correlation Matrix
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Tested', y='Mean Score', data=df)
plt.title('Relationship between Total Tested and Mean Score')
plt.xlabel('Total Tested')
plt.ylabel('Mean Score')
plt.show()
```

```python
# Example 9: Pair Plot - Relationships between Multiple Numerical Columns
selected_columns = ['Total Tested', 'Mean Score', 'Percent Scoring Below 65', 'Percent Scoring 65 or Above', 'Percent Scoring 80 or Above']
sns.pairplot(df[selected_columns])
plt.title('Pair Plot of Selected Columns')
plt.show()
```
